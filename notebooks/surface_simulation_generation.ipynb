```python
#Simulate n surfaces. Saves to the subfolder "simulated". Please create such subfolder first

# ? Minor design comment - maybe consider abstracting out some of the following code
# ? into a function that returns a 5-tuple (A, alpha, beta, B, delta) to avoid repetition?
# ? (applies to the next code block too)
# I agree, but since Igor wrote this script and given its complexity,
# I'm wary of changing anything about it in case something breaks
# or in case the seemingly duplicate code is actually different somehow

n=1000

for a in range(n):
    surface = np.zeros((256,256))
    A, alpha, beta, B, delta = np.random.random((5,))
    A = A * np.random.randint(1, 50 + 1)
    B = B * np.random.randint(10, 50 + 1)
    alpha = alpha * np.random.rand() / 8.
    beta = beta * np.random.rand() / 8.
    delta = delta * math.pi * 2
    
    A2, alpha2, beta2, B2, delta2 = np.random.random((5,))
    A2 = A2 * np.random.randint(10, 30 + 1)
    B2 = B2 * np.random.randint(10, 30 + 1)
    alpha2 = alpha2 * np.random.rand() / 15.
    beta2 = beta2 * np.random.rand() / 15.
    delta2 = delta2 * math.pi * 2
    
    A3, alpha3, beta3, B3, delta3 = np.random.random((5,))
    A3 = A3 * np.random.randint(10, 30 + 1)
    B3 = B3 * np.random.randint(10, 50 + 1)
    alpha3 = alpha3 * np.random.rand() / 4.
    beta3 = beta3 * np.random.rand() / 4.
    delta3 = delta3 * math.pi/2.1
    
    for i in range(256):
        for j in range(256):
            if (i==0 and j==0):
                surface[i][j] = 0
                pass
            if np.random.choice([True, False]):
                try:
                    surface[i][j] = abs((surface[i-1][j] + surface[i-1][j-1])/2.+ \
                    A3 * np.random.normal(scale = 2) + \
                    A * math.sin(alpha * i) * math.cos(beta * j + delta) + \
                     A2 * math.sin(alpha2 * i) * math.cos(beta2 * j + delta2) + \
                        A3 * math.sin(alpha3 * i) * math.cos(beta3 * j + delta3))
                except: 
                    surface[i][j] = abs((surface[i][j-1] + surface[i-1][j-1])/2.+ A3 * np.random.normal(scale = 2)+ \
                    A * math.sin(alpha * i) * math.cos(beta * j + delta) + \
                     A2 * math.sin(alpha2 * i) * math.cos(beta2 * j + delta2) + \
                        A3 * math.sin(alpha3 * i) * math.cos(beta3 * j + delta3))
            else:
                try:
                    # ? Alignment / indentation
                    # ? Will also probably help to explain what's going on here
                    # ? as well why you need a try/except
                    surface[i][j] = abs((surface[i][j-1] + surface[i-1][j-1])/2.+ A3 * np.random.normal(scale = 2)+ \
                    A * math.sin(alpha * i) * math.cos(beta * j + delta) + \
                     A2 * math.sin(alpha2 * i) * math.cos(beta2 * j + delta2) + \
                        A3 * math.sin(alpha3 * i) * math.cos(beta3 * j + delta3))
                except: 
                    surface[i][j] = abs((surface[i-1][j] + surface[i-1][j-1])/2.0+ \
                    A3 * np.random.normal(scale = 2) + \
                    A * math.sin(alpha * i) * math.cos(beta * j + delta) + \
                    A2 * math.sin(alpha2 * i) * math.cos(beta2 * j + delta2) + \
                    A3 * math.sin(alpha3 * i) * math.cos(beta3 * j + delta3))
                        
                        
    with open('simulated/test'+str(a)+'.asc','w') as myfile:
        myfile.write('# File Format = ASCII \n# Created by SPIP 5 \n# Original file: test0.asc \n# x-pixels = 256 \n# y-pixels = 256 \n# x-length = 5000 \n# y-length = 5000 \n# x-offset = 0 \n# y-offset = 0 \n# z-unit = nm \n# scanspeed = 50000 \n# forcecurve = 0 \n# voidpixels =0 \n# description =0: \n# Start of Data:\n')  
    f=open('simulated/test'+str(a)+'.asc','ab')
    np.savetxt(f,surface,delimiter='\t')
    f.close()
    
#     print(surface)  # just to see how a generated surface look like
#     plt.imshow(surface)
```
